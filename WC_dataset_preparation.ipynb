<h1>Getting event, freeze, related, tactics data

```python
match_ids = list(df_match['match_id'])

events = []
relateds = []
freezes = []
tactics = []
for id in match_ids:
  data = parser.event(id)

  events.append(data[0])
  relateds.append(data[1])
  freezes.append(data[2])
  tactics.append(data[3])

event_data = events[0]

for i in range(1, len(events)):
  event_data = pd.concat([event_data, events[i]], ignore_index=True)

related_data = relateds[0]

for i in range(1, len(relateds)):
  related_data = pd.concat([related_data, relateds[i]], ignore_index=True)

freeze_data = freezes[0]

for i in range(1, len(freezes)):
  freeze_data = pd.concat([freeze_data, freezes[i]], ignore_index=True)

tactic_data = tactics[0]

for i in range(1, len(tactics)):
  tactic_data = pd.concat([tactic_data, tactics[i]], ignore_index=True)
```
